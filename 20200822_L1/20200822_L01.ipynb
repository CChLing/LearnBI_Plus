### Action 1: 不用任何数学库，如何求出sqrt(10)，并且精确到小数点后10位

```python
# 二分查找法
def sqrt(num, precision=1e-10):
    if num == 0 or num == 1: # 特殊情况处理
        return num
    # 递归二分搜索
    def bi_search(low, high):
        mid = low + (high - low)/2
        if num - mid**2 > precision:
            return bi_search(mid, high)
        elif mid**2 - num > precision:
            return bi_search(low, mid)
        else:
            return mid
    return bi_search(0, num)
```

#### 以下基于课程代码：car_complain_bs.py

```python
# 根据request_url得到soup
def get_page_content(request_url):
    # 得到页面的内容
    headers={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36'}
    html=requests.get(request_url,headers=headers,timeout=10)
    content = html.text
    #print(content)

    # 通过content创建BeautifulSoup对象
    soup = BeautifulSoup(content, 'html.parser', from_encoding='utf-8')
    return soup
```

#### <font color=red>主要修改部分</font>

```python
# 分析当前页面的投诉
def analysis(BS):
    # 找到完整的投诉信息 table
    table = BS.find('table',class_="ar_c")
    # 创建DataFrame
    df = pd.DataFrame(columns = ['id', 'brand', 'car_model', 'type', 'desc', 'problem', 'datetime', 'status'])
    for i in range(len(table.find_all("tr"))):
        ll = []
        for td in table.find_all("tr")[i].find_all("td"):
            ll.append(td.text)
        if ll != []: # 屏蔽th行
            df.loc[i] = ll
    return df
```

```python
page_num = 20
base_url = 'http://www.12365auto.com/zlts/0-0-0-0-0-0_0-0-0-0-0-0-0-'

result = pd.DataFrame(columns = ['id', 'brand', 'car_model', 'type', 'desc', 'problem', 'datetime', 'status'])
for i in range(page_num):
    request_url = base_url+str(i+1)+'.shtml'
    soup = get_page_content(request_url)
    print(request_url)
    df = analysis(soup)
    print(df)
    result = result.append(df)
    
result.to_csv('car_complain.csv', index=False)
```
