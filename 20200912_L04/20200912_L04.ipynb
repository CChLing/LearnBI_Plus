### Thinking 1：奇异值分解SVD的原理是怎样的，都有哪些应用场景


原理： 常见情况下很多矩阵都是非对称的，可以通过将矩阵$A$转化为对称的方阵$AA^T$得到一个方阵，并且$AA^T$与$A^TA$是对称的方阵。$AA^T=P\Lambda _1P^T$,$A^TA=Q\Lambda _2Q^T$,$\Lambda _1$和$\Lambda _2$的非零特征值为$\sigma _1,\sigma _2,  \cdots , \sigma _k$($k\leqslant min(m,n)$)，此时矩阵$A$的特征值$\lambda _1=\sqrt{\sigma _1},\lambda _2=\sqrt{\sigma _2},\cdots ,\lambda _k=\sqrt{\sigma _k}$，可以求出$P$和$Q$得到奇异值分解$A=P\Lambda Q^T$。左奇异矩阵$P^{m\times m}$，右奇异矩阵$Q^{n\times n}$,$\Lambda $对角线上的非零元素为特征值$\lambda _1,\lambda _2,\cdots ,\lambda _k $。$A=P\Lambda Q^T=\lambda _1p_1q_1^T+\lambda _2p_2q_2^T+ \cdots +\lambda _kp_kq_k^T$


应用场景： 以上奇异值分解SVD实际上是无损重构了原矩阵$A$，而且奇异值在奇异值矩阵中按照从大到小排列，并且在很多情况下，前10%甚至前1%的奇异值的和就占了全部奇异值之和的90%以上，也就是说，可以用最大的k个奇异值和对应的左右奇异向量来近似还原原矩阵。  
在实际应用场景下，主成分分析（PCA）在数据量较大的情况下PCA算法可以不做特征分解，转为用SVD来减少计算量。  
SCD也可以用于简单的图像压缩，其优点是简化数据，去除噪声。

### Thinking 2：funkSVD, BiasSVD，SVD++算法之间的区别是怎样的


funkSVD：用户-评分矩阵$R$通常是很稀疏的，原始SVD无法使用，因此，假设$R^{m\times n}\approx P^{k\times m}Q^{k\times n}$, 转换为最优化问题：
$$\underset{P,Q}{min}L = \sum_{(i,j)\in K}(R_{i,j}-p_i^Tq_j)^2+\lambda(\sum_{i=1}^{m}\left | p_i \right |^2 + \sum_{j=1}^{n}\left | q_j \right |^2)$$
可用随机梯度下降法（SGD）优化模型参数。
BiasSVD：在funkSVD的基础上，加入用户偏好和商品偏好，此时评分$R_{i,j}=\mu +b_i+b_j+p_i^Tq_j$，即最优化问题：
$$\underset{P,Q,b_i,b_j}{min}L = \sum_{(i,j)\in K}(R_{i,j}-\mu -b_i-b_j-p_i^Tq_j)^2+\lambda (\sum_{i=1}^{m}\left | b_i \right |^2 + \sum_{j=1}^{n}\left | b_j \right |^2 +\sum_{i=1}^{m}\left | p_i \right |^2 + \sum_{j=1}^{n}\left | q_j \right |^2)$$
可用随机梯度下降法（SGD）优化模型参数。
SVD++： 在BiasSVD的基础上，考虑用户的隐式反馈。最优化问题：
$$\underset{P,Q,b_i,b_j}{min}L = \sum_{(i,j)\in K}(R_{i,j}-\mu -b_i-b_j-p_i^Tq_j-q_j^T \left | I(i) \right |^{-\frac{1}{2}}\sum _{s\in I(i)}y_s)^2+\lambda (\sum_{i=1}^{m}\left | b_i \right |^2 + \sum_{j=1}^{n}\left | b_j \right |^2 +\sum_{i=1}^{m}\left | p_i \right |^2 + \sum_{j=1}^{n}\left | q_j \right |^2 + \sum _{s\in I(i)}\left | y_s \right |)$$
可用随机梯度下降法（SGD）优化模型参数。

### Thinking 4：假设一个小说网站，有N部小说，每部小说都有摘要描述。如何针对该网站制定基于内容的推荐系统，即用户看了某部小说后，推荐其他相关的小说。原理和步骤是怎样的


可以将N部小说的摘要做分词然后去除停用词，进行n—Gram词元抽取（可以同时用unigram，bigram，trigram），使用TF-IDF提取文本特征，然后计算其他小说与用户看过的小说内容的余弦相似度，将相似度较高的小说推荐给用户。原理是用户看过的小说通常代表了用户的喜好，通过以上操作可以计算不同小说之间在内容上的相似度，推荐在内容上与用户看过的小说相近的小说，用户很大概率会喜欢。

### Thinking 5：Word2Vec的应用场景有哪些


Word2Vec能够计算得到各个词之间的相似度，所以可以基于内容相似度来进行推荐，比如新闻或文章推荐。  
Word2Vec还潜在包含词与词之间的关系，比如“男人”和“女人”的关系与“皇帝”与“皇后”的关系相近在Word2Vec中表现为$V(男人)-V(女人)$与$V(皇帝)-V(皇后)$的余弦相似度很大，因此在NLP（自然语言处理）中常用文本中单词的Word2Vec作为单词的表示来进行文本信息的处理。

```python
from scipy.linalg import svd
import matplotlib.pyplot as plt
```

```python
from PIL import Image
import numpy as np
# 加载图片
image = Image.open('data/256.bmp')
A = np.array(image)
# 显示原图像
plt.imshow(A, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
A.shape
```

```python
# 取前k个特征，对图像进行还原
def get_image_feature(s, k):
    # 对于S，只保留前K个特征值
    s_temp = np.zeros(s.shape[0])
    s_temp[0:k] = s[0:k]
    s = s_temp * np.identity(s.shape[0])
    # 用新的s_temp，以及p,q重构A
    temp = np.dot(p,s)
    temp = np.dot(temp,q)
    plt.imshow(temp, cmap=plt.cm.gray, interpolation='nearest')
    plt.show()
    print(A-temp)
```

```python
# 对图像矩阵A进行奇异值分解，得到p,s,q
p,s,q = svd(A, full_matrices=False)
# 取前k个特征，对图像进行还原
get_image_feature(s, 5)
get_image_feature(s, 50)
get_image_feature(s, 500)
```

<b>文件分割</b>

```python
# 对文件进行分词
import jieba
import os
from utils import files_processing
```

```python
# 源文件所在目录
source_folder = 'data/three_kingdoms/source'
segment_folder = 'data/three_kingdoms/segment'
```

```python
# 字词分割，对整个文件内容进行字词分割
def segment_lines(file_list,segment_out_dir,stopwords=[]):
    for i,file in enumerate(file_list):
        segment_out_name=os.path.join(segment_out_dir,'segment_{}.txt'.format(i))
        with open(file, 'rb') as f:
            document = f.read()
            document_cut = jieba.cut(document)
            sentence_segment=[]
            for word in document_cut:
                if word not in stopwords:
                    sentence_segment.append(word)
            result = ' '.join(sentence_segment)
            result = result.encode('utf-8')
            with open(segment_out_name, 'wb') as f2:
                f2.write(result)
```

```python
# 加载停用词
with open("data/chinese_stopwords.txt", "r", encoding ="utf-8") as f:
    stopwords = [i[:-1] for i in f.readlines()]
```

```python
# 对source中的txt文件进行分词，输出到segment目录中
file_list=files_processing.get_files_list(source_folder, postfix='*.txt')
segment_lines(file_list, segment_folder, stopwords)
```
